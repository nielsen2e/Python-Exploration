## Exploratory Data Analysis

**Let's explore the data!**


#### Highest grossing movie per year

```python
CompanySum = data.groupby(['company', 'year'])[["gross"]].sum()

CompanySorted = CompanySum.sort_values(['gross', 'company', 'year'], ascending = False)[:10]

CompanySorted
```

#### The name of the Highest grossing movie

```python
data[data['gross'] == data['gross'].max()]
```

#### The name of the Lowest grossing movie

```python
data[data['gross'] == data['gross'].min()]
```

#### What movies did Robert Downey Jr. act star in ?

```python
data[data['star'] == 'Robert Downey Jr.'][['name', 'year']]
```

#### What are the top 5 most common companies?

```python
data['company'].value_counts().head()
```

```python
# Let's check for missing data 
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
        
```

```python
#Let's look at all the data

pd.set_option('display.max_rows', None)
```

```python
# Let's drop any duplicates in company column

# data.drop_duplicates() -- To drop the values of all the columns
# We are checking to see if there are issues in the data e.g walt disney changed name during different times
# due to different timeframes

data['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
# Let's check for columns correlated to gross revenue
# Looking at the ones with high correlation
# Budget has a high correlation
# Company name high correlation
```

```python
# Lets compare budget and gross revenue using scatter plot

plt.scatter(x = data['budget'], y = data['gross'])

plt.title('Budget vs Gross Revenue')

plt.xlabel('Budget for Movies')
plt.ylabel('Gross Revenue')

plt.show
```

```python
#Lets do a regression plot to check for correlation
# This shows a postive correlation because its going up

sns.regplot(x = 'budget', y = 'gross', data=data, scatter_kws = {'color':'red'}, line_kws = {'color':'blue'})
```

```python
# Lets start looking at correlation
# correlation works on numerical values
#methods of correlation; pearson, kendall, spearman
data.corr(method = 'pearson')
```

```python
# High correlation between budget and gross
#Lets vizualize the correlation matrix

correlation_matrix = data.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)


plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
# lets look at Company
# Lets change company to a numeric feature to be able to get more correlations

data.head()
```

```python
df_numerized = data
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
data
```

```python
#Lets vizualize the numerized correlation matrix

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)


plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
# Lets organize the data to see the ones with the highest corelation
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# Lets sort the pairs
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.4]
                         
high_corr
```
